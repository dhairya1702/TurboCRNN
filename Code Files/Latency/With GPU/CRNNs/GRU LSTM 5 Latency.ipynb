```python
import time
```

```python
model_run_times= []
```

```python
def compute_latency(test_audios):
  for i in test_audios:
    start = time.time()
    model.predict(np.expand_dims(i, axis=0))
    end = time.time()
    model_run_times.append(end - start)
```

```python
model_run_times= []
compute_latency(X_test)
```

```python
print("GRULSTM-WithoutAttention-WithDense.h5\nAverage Model Latency: %.2f ms" % round(np.mean(model_run_times)*1000,2))
```
